```python
import cv2
import numpy as np
import random
from matplotlib import pyplot as plt

def random_crop(image, crop_size):
    """随机裁剪图像"""
    h, w = image.shape[:2]
    crop_h, crop_w = crop_size
    
    # 确保裁剪尺寸不超过原图尺寸
    if crop_h > h or crop_w > w:
        raise ValueError("裁剪尺寸不能大于原图尺寸")
    
    # 随机选择裁剪位置
    x = random.randint(0, w - crop_w)
    y = random.randint(0, h - crop_h)
    
    # 执行裁剪
    cropped_image = image[y:y+crop_h, x:x+crop_w]
    return cropped_image

def random_horizontal_flip(image, p=0.5):
    """随机水平翻转图像"""
    # 根据概率决定是否翻转
    if random.random() < p:
        flipped_image = cv2.flip(image, 1)  # 1表示水平翻转
        return flipped_image
    return image

def random_scale(image, min_scale=0.5, max_scale=2.0):
    """多尺度随机缩放图像"""
    # 随机选择缩放比例
    scale = random.uniform(min_scale, max_scale)
    
    # 计算新的尺寸
    h, w = image.shape[:2]
    new_h, new_w = int(h * scale), int(w * scale)
    
    # 执行缩放
    scaled_image = cv2.resize(image, (new_w, new_h), interpolation=cv2.INTER_LINEAR)
    return scaled_image

def visualize_augmentations(image_path):
    """可视化各种数据增强效果"""
    # 读取图像
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # 转换为RGB格式
    
    # 原图
    plt.figure(figsize=(15, 10))
    plt.subplot(2, 2, 1)
    plt.title('Original Image')
    plt.imshow(image)
    
    # 随机裁剪
    cropped = random_crop(image, (image.shape[0]//2, image.shape[1]//2))
    plt.subplot(2, 2, 2)
    plt.title('Random Crop')
    plt.imshow(cropped)
    
    # 随机水平翻转
    flipped = random_horizontal_flip(image)
    plt.subplot(2, 2, 3)
    plt.title('Random Horizontal Flip')
    plt.imshow(flipped)
    
    # 随机缩放
    scaled = random_scale(image)
    plt.subplot(2, 2, 4)
    plt.title('Random Scale')
    plt.imshow(scaled)
    
    plt.tight_layout()
    plt.show()

# 使用示例
if __name__ == "__main__":
    image_path = "/home/featurize/work/MMSegmentation_Tutorials-main/project/肾小球切片语义分割/Glomeruli-dataset/images/SAS_21883_001_104.png"  # 替换为你的图像路径
    visualize_augmentations(image_path)
```
